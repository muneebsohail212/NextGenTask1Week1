```python
import pandas as pd
import numpy as np
```

```python
from faker import Faker
import random

fake = Faker()
```

```python
domains = [
    "Web Development",
    "AI",
    "Graphic Design",
    "Data Science",
    "Cyber Security"
]

status = [
    "Selected",
    "Rejected",
    "Under Review"
]

rows=[]

for i in range(100):
    rows.append({
        "Applicant Name":fake.name(),
        "Email":fake.email(),
        "Phone":fake.phone_number(),
        "Domain Applied":random.choice(domains),
        "University":fake.company(),
        "Application Date":fake.date_between("-120d","today"),
        "Status":random.choice(status)
    })

df=pd.DataFrame(rows)
```

```python
df.loc[5,"Email"]=None
df.loc[8,"University"]=None
df.loc[10,"Status"]=None
df.loc[15,"Applicant Name"]=None

df.loc[20,"Domain Applied"]="web dev"
df.loc[21,"Domain Applied"]="WEB DEV"
df.loc[22,"Domain Applied"]="web development"

df.loc[30,"Application Date"]="12/05/2026"
df.loc[31,"Application Date"]="2026/06/01"

df=df._append(df.iloc[4],ignore_index=True)
df=df._append(df.iloc[9],ignore_index=True)
```

```python
df.to_csv("applicants.csv",index=False)
```

```python
df=pd.read_csv("applicants.csv")
```

```python
df.shape
df.head()
df.info()
df.describe(include="all")
df.isnull().sum()
df.duplicated().sum()
```

handling missing values now

```python
df=df.dropna(subset=["Applicant Name"])
```

```python
df["Email"]=df["Email"].fillna("not_provided@example.com")
```

```python
df["University"]=df["University"].fillna("Not Specified")
```

```python
df["Status"]=df["Status"].fillna("Under Review")
```

applicant name is critical so i remove it university was not so i make it not specified

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df=df.drop_duplicates(subset=["Email"])
```

for spaces remove

```python
df["Applicant Name"]=df["Applicant Name"].str.strip()

df["Domain Applied"]=df["Domain Applied"].str.strip()

df["Status"]=df["Status"].str.strip()
```

for lower case

```python
df["Domain Applied"]=df["Domain Applied"].str.lower()
```

```python
mapping={
"web dev":"Web Development",
"web development":"Web Development",
"WEB DEV":"Web Development",
"ai":"AI",
"graphic design":"Graphic Design",
"data science":"Data Science",
"cyber security":"Cyber Security"
}
```

```python
df["Domain Applied"]=df["Domain Applied"].replace(mapping)
```

```python
df["Domain Applied"].unique()
```

```python
df["Status"]=df["Status"].str.title()
```

```python
df["Application Date"]=pd.to_datetime(
    df["Application Date"],
    errors="coerce"
)
```

```python
df["Phone"]=(
    df["Phone"]
    .astype(str)
    .str.replace("-","",regex=False)
    .str.replace(" ","",regex=False)
)
```
